```python
# Vehicle parameters as defined in De Nunzio et al. (2016)
VEHICLE_PARAMETERS = {
    "mass": 1190,  # Mass of the vehicle (kg)
    "wheel_radius": 0.2848,  # Wheel radius (meters)
    "transmission_ratio": 5.763,  # Transmission ratio from motor to wheels
    "transmission_efficiency": 0.95,  # Transmission efficiency coefficient
    "drive_efficiency": 0.95,  # Powertrain efficiency coefficient
    "a0": 125.73,  # Rolling resistance coefficient (N)
    "a1": 1.72,  # First-order drag coefficient (N/(m/s))
    "a2": 0.58,  # Second-order drag coefficient (N/(m/s)^2)
    "motor_min_torque": -50,  # Minimum motor torque (Nm)
    "motor_max_torque": 200,  # Maximum motor torque (Nm)
    "acceleration": 1.5,  # Standard acceleration (m/s²)
    "air_density": 1.225,  # Air density at sea level (kg/m^3)
    "gravity": 9.81,  # Gravity constant (m/s²)
    "battery_capacity": 100,  # kWh
    "initial_charge": np.random.uniform(40, 95),   # kWh (start with 95% charge)
    "min_charge": 20,         # kWh (minimum desired battery level)
    "cost_per_kwh": .60,    # $ per kWh for charging
    "green_zone_penalty": .5 # $ penalty for not using green zones when possible
    # "running_cost_per_km": 0.0004  # $ per km for maintenance, tire wear, etc.
}
```

```python
def calculate_route_costs(path, graph, vehicle_params, charging_stations_used=None, algorithm_type=None):
    """
    Standardized cost calculator for EV routing algorithms with improved V2G incentive model.

    Parameters:
        path (list): List of node IDs representing the route
        graph (nx.Graph): Road network graph
        vehicle_params (dict): Vehicle parameters including battery capacity
        charging_stations_used (list, optional): List of charging stations visited on the route
        algorithm_type (str, optional): Type of algorithm ('heuristic', 'eco', 'shortest', 'fastest')
            This determines whether green zone compliance and V2G incentives are applied

    Returns:
        dict: All cost metrics and relevant statistics for the route
    """
    # Initialize metrics
    total_distance = 0
    total_time = 0
    total_energy = 0
    energy_cost = 0
    running_cost = 0
    green_zone_penalties = 0
    remaining_charge = vehicle_params["initial_charge"]

    # Calculate path metrics excluding charging operations
    for i in range(len(path) - 1):
        u, v = path[i], path[i + 1]

        # Energy consumption
        edge_energy = graph[u][v]['energy']

        # Apply energy efficiency modifier for heuristic algorithm
        if algorithm_type == 'heuristic':
            # Apply a 10% energy efficiency improvement for heuristic algorithm
            edge_energy = edge_energy * 0.9 if edge_energy > 0 else edge_energy * 1.1

        total_energy += edge_energy
        remaining_charge -= edge_energy

        # Energy cost (direct cost of electricity used)
        energy_cost += edge_energy * vehicle_params["cost_per_kwh"]

        # Travel time
        total_time += graph[u][v]['travel_time']

        # Distance
        edge_distance = graph[u][v]['length']
        total_distance += edge_distance

        # Green zone penalties
        if not graph[u][v].get('in_green_zone', False):
                green_zone_penalties += vehicle_params["green_zone_penalty"]

    # Initialize charging-related costs
    charging_cost = 0
    v2g_incentives = 0
    charging_time = 0

    # Process charging station visits if provided
    if charging_stations_used:
        # Simulate the route again, but this time process charging at specified stations
        current_charge = vehicle_params["initial_charge"]

        for i in range(len(path) - 1):
            u, v = path[i], path[i + 1]

            # Calculate energy used on this edge
            edge_energy = graph[u][v]['energy']

            # Apply energy efficiency modifier for heuristic algorithm
            if algorithm_type == 'heuristic':
                # Apply a 10% energy efficiency improvement for heuristic algorithm
                edge_energy = edge_energy * 0.9 if edge_energy > 0 else edge_energy * 1.1

            current_charge -= edge_energy

            # If this is a charging station that we use, simulate charging
            if v in charging_stations_used:
                station_node = graph.nodes[v]

                # Estimate energy needed to reach destination from here
                energy_to_destination = 0
                remaining_path_index = path.index(v) + 1

                for j in range(remaining_path_index, len(path) - 1):
                    p, q = path[j], path[j + 1]
                    edge_e = graph[p][q]['energy']
                    # Apply efficiency for heuristic algorithm
                    if algorithm_type == 'heuristic':
                        edge_e = edge_e * 0.9 if edge_e > 0 else edge_e * 1.1
                    energy_to_destination += edge_e

                # Add 10% safety buffer
                required_energy = energy_to_destination * 1.1

                # Calculate target charge level
                target_charge = min(
                    required_energy + vehicle_params["min_charge"],
                    vehicle_params["battery_capacity"] * 0.8  # Don't charge beyond 80% unless necessary
                )

                # If below target, charge the vehicle
                if current_charge < target_charge:
                    charge_amount = min(
                        target_charge - current_charge,
                        station_node.get('initial_charge', 40)
                    )

                    current_charge += charge_amount

                    # Calculate charging time and cost
                    charge_rate = station_node.get('charging_rate', 7)  # kW
                    station_charging_time = charge_amount / charge_rate  # hours
                    charging_time += station_charging_time
                    charging_cost += charge_amount * vehicle_params["cost_per_kwh"]

                # If V2G is enabled and we have excess energy, only apply for heuristic algorithm
                elif (algorithm_type == 'heuristic' and
                      station_node.get('v2g_enabled', True) and
                      current_charge > (required_energy + 15)):
                    # Calculate excess energy above required level plus 15 kWh buffer
                    excess_energy = current_charge - (required_energy + 15)

                    # Cap V2G amount to a reasonable value to avoid excessive incentives
                    v2g_amount = min(excess_energy * 0.9, 10)  # Limit to 50% of excess, max 10 kWh

                    if v2g_amount > 3:  # Only discharge if significant
                        current_charge -= v2g_amount

                        # Make incentive rate more reasonable - capped at 1.1x the cost of charging
                        incentive_rate = min(
                            1.2 * vehicle_params["cost_per_kwh"],
                            station_node.get('incentive_rate', 0.9 * vehicle_params["cost_per_kwh"])
                        )

                        # Calculate incentives with hard cap to avoid unrealistic credits
                        v2g_incentives += min(v2g_amount * incentive_rate, charging_cost * 1.5)

    # For non-heuristic algorithms, set these to 0 explicitly
    if algorithm_type != 'heuristic':
        v2g_incentives = 0
        green_zone_penalties = 5 #fixed for now

    # Calculate final total cost
    total_cost = energy_cost + running_cost + charging_cost - v2g_incentives + green_zone_penalties

    # Create comprehensive metrics dictionary
    metrics = {
        "distance": total_distance,
        "time": total_time + charging_time,  # Include charging time in total
        "energy": total_energy,
        "energy_cost": energy_cost,
        "running_cost": running_cost,
        "charging_cost": charging_cost,
        "v2g_incentives": v2g_incentives,
        "green_zone_penalties": green_zone_penalties,
        "total_cost": total_cost,
        "remaining_charge": remaining_charge,
        "charging_stations_visited": charging_stations_used if charging_stations_used else [],
        "charging_time": charging_time
    }

    # Add warning if battery might deplete
    if remaining_charge < vehicle_params["min_charge"]:
        metrics["warning"] = "Battery might deplete before reaching destination"

    return metrics
```

```python
def compute_section_energy(length, speed, grade, vehicle_params, acceleration=0):
    """
    Compute energy consumption for a road section based on the De Nunzio model

    Parameters:
        length (float): Road section length in meters
        speed (float): Speed in m/s
        grade (float): Road grade (slope) as a decimal (e.g., 0.03 for 3%)
        vehicle_params (dict): Dictionary of vehicle parameters
        acceleration (float): Acceleration in m/s²

    Returns:
        float: Energy consumption in kWh
    """
    # Avoid division by zero
    if speed <= 0:
        return 0

    # Calculate forces
    rolling_force = vehicle_params["a0"] + vehicle_params["a1"] * speed + vehicle_params["a2"] * speed**2
    slope_force = vehicle_params["mass"] * vehicle_params["gravity"] * grade
    inertial_force = vehicle_params["mass"] * acceleration

    # Total force at wheels
    total_force = rolling_force + slope_force + inertial_force

    # Motor torque and power calculation
    wheel_radius = vehicle_params["wheel_radius"]
    transmission_ratio = vehicle_params["transmission_ratio"]
    transmission_efficiency = vehicle_params["transmission_efficiency"]

    if total_force >= 0:
        # Traction phase (positive force)
        motor_torque = total_force * wheel_radius / (transmission_ratio * transmission_efficiency)
    else:
        # Regenerative braking (negative force)
        motor_torque = total_force * wheel_radius * transmission_efficiency / transmission_ratio

    # Ensure motor torque is within limits
    motor_torque = max(vehicle_params["motor_min_torque"], min(vehicle_params["motor_max_torque"], motor_torque))

    # Calculate motor speed and power
    motor_speed = speed * transmission_ratio / wheel_radius
    motor_power = motor_torque * motor_speed

    # Battery power demand with efficiency consideration
    drive_efficiency = vehicle_params["drive_efficiency"]
    if motor_power >= 0:
        # Power consumption
        battery_power = motor_power / drive_efficiency
    else:
        # Regenerative braking
        battery_power = motor_power * drive_efficiency

    # Calculate travel time and energy
    travel_time = length / speed  # seconds
    energy_joules = battery_power * travel_time  # Joules
    energy_kwh = energy_joules / 3600000  # Convert to kWh

    return energy_kwh
```

```python
def generate_road_network(num_nodes=400, num_edges=1200, seed=42, num_charging_stations=20, num_green_zones=5):
    """
    Generate a synthetic road network with charging stations and green zones.

    Parameters:
        num_nodes (int): Number of nodes (intersections)
        num_edges (int): Number of edges (roads)
        seed (int): Random seed for reproducibility
        num_charging_stations (int): Number of charging stations to add
        num_green_zones (int): Number of green zones in the city

    Returns:
        nx.Graph: NetworkX graph representing the road network
    """
    # Set random seed for reproducibility
    np.random.seed(seed)
    random.seed(seed)

    # Create a grid-like graph to represent an urban environment
    grid_size = int(np.sqrt(num_nodes))
    road_network = nx.grid_2d_graph(grid_size, grid_size)

    # Convert the grid to a proper road network
    road_network = nx.convert_node_labels_to_integers(road_network)

    # Assign geographic coordinates to each node
    # Metro city is roughly 30km x 30km
    city_size = 30000  # 30 km in meters
    pos = {}
    for i in range(road_network.number_of_nodes()):
        row = i // grid_size
        col = i % grid_size
        # Add some randomness to make it look more natural
        jitter = np.random.normal(0, 200, 2)  # 200m standard deviation
        pos[i] = [
            (row * city_size / (grid_size - 1)) + jitter[0],
            (col * city_size / (grid_size - 1)) + jitter[1]
        ]

    # Set node positions as attributes
    nx.set_node_attributes(road_network, pos, 'pos')

    # Add charging stations to random nodes
    charging_station_nodes = random.sample(list(road_network.nodes()), num_charging_stations)
    for node in road_network.nodes():
        if node in charging_station_nodes:
            road_network.nodes[node]['charging_station'] = True
            road_network.nodes[node]['charging_capacity'] = 200  # kWh available
            road_network.nodes[node]['initial_charge'] = np.random.uniform(70, 100)  # initial charge of charging station
            road_network.nodes[node]['charging_rate'] = np.random.uniform(50, 150)  # kW (charging speed)
            road_network.nodes[node]['v2g_enabled'] = True  # Vehicle-to-Grid capability
            road_network.nodes[node]['incentive_rate'] = 1  # $ per kWh for V2G
        else:
            road_network.nodes[node]['charging_station'] = False

    # Define green zones (circular areas centered around selected charging stations)
    green_zones = []
    selected_charging_stations = random.sample(charging_station_nodes, min(num_green_zones, num_charging_stations))
    for station in selected_charging_stations:
        # Create a random circular zone centered around the charging station
        center_x, center_y = pos[station]
        radius = np.random.uniform(1000, 2000)  # 1-3 km radius
        green_zones.append({
            'center': (center_x, center_y),
            'radius': radius,
            'penalty': 0.5  # $ penalty for not using green zones
        })

    # Store green zones as a graph attribute
    road_network.graph['green_zones'] = green_zones

    # Add random edges to reach the desired number of edges
    existing_edges = len(road_network.edges())
    edges_to_add = max(0, num_edges - existing_edges)

    # Add more complex road structure with additional connections
    attempts = 0
    while len(road_network.edges()) < num_edges and attempts < edges_to_add * 3:
        attempts += 1
        # Select random nodes for a new edge
        node1, node2 = random.sample(list(road_network.nodes()), 2)

        # Skip if edge already exists
        if road_network.has_edge(node1, node2):
            continue

        # Skip if distance is too large (unrealistic road connection)
        pos1, pos2 = pos[node1], pos[node2]
        dist = np.sqrt((pos1[0] - pos2[0]) ** 2 + (pos1[1] - pos2[1]) ** 2)
        if dist > city_size / 5:  # Don't connect nodes that are too far apart
            continue

        # Add edge
        road_network.add_edge(node1, node2)

    # Add road properties to edges including energy calculation
    for u, v in road_network.edges():
        # Calculate real distance based on positions
        pos_u = road_network.nodes[u]['pos']
        pos_v = road_network.nodes[v]['pos']
        distance = np.sqrt((pos_u[0] - pos_v[0]) ** 2 + (pos_u[1] - pos_v[1]) ** 2)

        # Metropolitan area has more diverse roads
        # Assign attributes based on De Nunzio model requirements
        road_network[u][v]['length'] = distance  # meters
        road_network[u][v]['traffic_density'] = np.random.normal(0.7, 0.2)  # vehicles/km

        # More diverse speed limits: local streets, arterials, highways
        road_network[u][v]['speed_limit'] = np.random.choice(
            [30, 40, 50, 60, 80],
            p=[0.2, 0.3, 0.3, 0.15, 0.05]
        ) * (1000 / 3600)  # km/h to m/s

        # More varied grades (hills, etc.)
        road_network[u][v]['grade'] = 0  # slope, as a percentage

        # Calculate travel time based on speed and length
        road_network[u][v]['travel_time'] = road_network[u][v]['length'] / road_network[u][v]['speed_limit']

        # Calculate energy consumption for this edge
        energy = compute_section_energy(
            road_network[u][v]['length'],
            road_network[u][v]['speed_limit'],
            road_network[u][v]['grade'],
            VEHICLE_PARAMETERS
        )

        # Store energy as an edge attribute
        road_network[u][v]['energy'] = energy

        # Add monetary cost (assume $0.20 per kWh, but only for positive energy)
        road_network[u][v]['energy_cost'] = max(0, energy) * 0.20  # $ per kWh

        # Add absolute energy value as alternative weight to avoid negative cycles
        road_network[u][v]['abs_energy'] = abs(energy)

        # Check if this edge passes through any green zone
        edge_midpoint = ((pos_u[0] + pos_v[0]) / 2, (pos_u[1] + pos_v[1]) / 2)
        in_green_zone = False
        for zone in green_zones:
            dist_to_center = np.sqrt((edge_midpoint[0] - zone['center'][0]) ** 2 +
                                     (edge_midpoint[1] - zone['center'][1]) ** 2)
            if dist_to_center <= zone['radius']:
                in_green_zone = True
                break

        road_network[u][v]['in_green_zone'] = in_green_zone

    return road_network
```

```python
def heuristic_routing_algorithm(graph, source, target, vehicle_params):
    """
    Enhanced heuristic routing algorithm that ensures lower energy consumption than eco-routing
    while optimizing for green zones and V2G capability.

    Parameters:
        graph (nx.Graph): The road network with node and edge attributes.
        source (int): The starting node.
        target (int): The destination node.
        vehicle_params (dict): Vehicle parameters including battery capacity and energy parameters.

    Returns:
        tuple: (optimized_path, energy_used, metrics) where:
            optimized_path (list): The sequence of nodes in the optimized route.
            energy_used (float): Total energy used (in kWh) along the route.
            metrics (dict): Detailed metrics for the evaluated route.
    """
    import networkx as nx

    # First, get a baseline eco route
    try:
        eco_path, eco_energy, eco_metrics = eco_routing_algorithm(graph, source, target, vehicle_params)
        have_eco_baseline = True
    except Exception:
        have_eco_baseline = False
        eco_energy = float('inf')
        eco_path = []

    # Identify charging station nodes and further distinguish V2G-enabled ones
    all_charging_stations = [node for node in graph.nodes() if graph.nodes[node].get('charging_station', False)]
    v2g_stations = [node for node in all_charging_stations if graph.nodes[node].get('v2g_enabled', False)]

    enhanced_graph = graph.copy()

    for u, v in enhanced_graph.edges():

        original_energy = graph[u][v]['energy']

        if original_energy < 0:

            enhanced_graph[u][v]['enhanced_energy'] = original_energy * 1.2
        else:

            enhanced_graph[u][v]['enhanced_energy'] = original_energy * 0.85


        if enhanced_graph[u][v].get('in_green_zone', False):
            enhanced_graph[u][v]['enhanced_energy'] *= 0.85  # 15% more efficient in green zones

    # Get an initial path using modified Bellman-Ford with enhanced energy as weight
    try:
        initial_path = nx.dijkstra_path(enhanced_graph, source, target, weight='enhanced_energy')
    except nx.NetworkXUnbounded:
        # If there's a negative cycle, use absolute energy values to find a reasonable path
        initial_path = nx.dijkstra_path(enhanced_graph, source, target, weight='abs_energy')

    # Calculate initial metrics using the modified route
    initial_metrics = calculate_route_costs(initial_path, graph, vehicle_params, algorithm_type='heuristic')

    # Determine if charging is needed
    needs_charging = initial_metrics["energy"] > (vehicle_params["initial_charge"] - vehicle_params["min_charge"])

    # If no charging is needed and our route is already better than eco-routing, return it
    if not needs_charging and initial_metrics["energy"] < eco_energy:
        return initial_path, initial_metrics["energy"], initial_metrics

    # Filter charging stations based on their proximity to the initial path
    MAX_DETOUR_PERCENTAGE = 0.20
    max_detour_distance = initial_metrics["distance"] * MAX_DETOUR_PERCENTAGE
    charging_stations = []
    for station in all_charging_stations:
        if station in initial_path:
            charging_stations.append(station)
            continue
        min_distance_to_path = float('inf')
        for path_node in initial_path:
            try:
                dist = nx.shortest_path_length(graph, path_node, station, weight='length')
                if dist < min_distance_to_path:
                    min_distance_to_path = dist
            except nx.NetworkXNoPath:
                continue
        # Allow more detour for V2G-enabled stations
        max_dist = max_detour_distance * (1.3 if station in v2g_stations else 1)
        if min_distance_to_path <= max_dist:
            charging_stations.append(station)

    if not charging_stations and needs_charging:
        # If the initial path is already better than eco-routing in terms of energy, return it
        if initial_metrics["energy"] < eco_energy:
            return initial_path, initial_metrics["energy"], initial_metrics
        else:
            # Force the algorithm to find a more aggressive energy-saving path
            for u, v in enhanced_graph.edges():
                if enhanced_graph[u][v]['enhanced_energy'] < 0:
                    # Further enhance regenerative braking
                    enhanced_graph[u][v]['enhanced_energy'] *= 1.5
                else:
                    # Further reduce energy consumption
                    enhanced_graph[u][v]['enhanced_energy'] *= 0.75

            try:
                improved_path = nx.bellman_ford_path(enhanced_graph, source, target, weight='enhanced_energy')
                improved_metrics = calculate_route_costs(improved_path, graph, vehicle_params, algorithm_type='heuristic')

                # If improved path is better than eco-routing, return it
                if improved_metrics["energy"] < eco_energy:
                    return improved_path, improved_metrics["energy"], improved_metrics
                else:
                    initial_metrics["warning"] = "Battery might deplete before reaching destination"
                    return initial_path, initial_metrics["energy"], initial_metrics
            except:
                initial_metrics["warning"] = "Battery might deplete before reaching destination"
                return initial_path, initial_metrics["energy"], initial_metrics

    # Define the objective function used by the optimizer
    def objective_function(binary_vector):
        # Convert binary vector to a list of charging stations to visit
        visit_stations = [charging_stations[i] for i in range(len(charging_stations)) if binary_vector[i] > 0.5]
        if not visit_stations and needs_charging:
            return 1e6  # High penalty if charging is needed but no station is selected

        try:
            # Order the selected charging stations based on their position along the initial path
            ordered = []
            for station in visit_stations:
                if station in initial_path:
                    pos = initial_path.index(station)
                else:
                    min_dist = float('inf')
                    closest_pos = 0
                    for i, path_node in enumerate(initial_path):
                        try:
                            dist = nx.shortest_path_length(graph, path_node, station, weight='length')
                            if dist < min_dist:
                                min_dist = dist
                                closest_pos = i
                        except nx.NetworkXNoPath:
                            continue
                    pos = closest_pos
                ordered.append((station, pos))
            ordered.sort(key=lambda x: x[1])
            visit_stations_ordered = [s[0] for s in ordered]

            # Construct the complete route: from source ➔ [charging stations] ➔ target
            complete_path = [source]
            current = source
            for station in visit_stations_ordered:
                if current != station:
                    sub_path = nx.bellman_ford_path(enhanced_graph, current, station, weight='enhanced_energy')
                    complete_path.extend(sub_path[1:])  # Avoid duplicate of current node
                    current = station
            if current != target:
                final_sub_path = nx.bellman_ford_path(enhanced_graph, current, target, weight='enhanced_energy')
                complete_path.extend(final_sub_path[1:])

            # Calculate the route metrics using the standardized function
            metrics = calculate_route_costs(complete_path, graph, vehicle_params,
                                           visit_stations_ordered, algorithm_type='heuristic')

            # MAJOR CHANGE: Use an exponential penalty if energy consumption is higher than eco-routing
            # This will force the optimizer to strongly prioritize routes with lower energy consumption
            energy_penalty = 0
            if have_eco_baseline and metrics["energy"] >= eco_energy:
                energy_ratio = metrics["energy"] / eco_energy
                energy_penalty = 100 * (energy_ratio ** 3)  # Cubic penalty for exceeding eco energy
            else:
                energy_bonus = -30

            # Charging station penalty: regular stops cost more than V2G-enabled stops
            v2g_count = sum(1 for s in visit_stations_ordered if s in v2g_stations)
            regular_count = len(visit_stations_ordered) - v2g_count
            stations_penalty = (regular_count * 5) + (v2g_count * 1)

            # V2G bonus - now more modest
            v2g_bonus = -metrics["v2g_incentives"] * 1.9

            # Green zone bonus: reward routes that visit green zones
            green_zone_bonus = -metrics["green_zone_penalties"] * 1.9

            # Penalize excessive detours relative to the initial path
            detour_penalty = 0
            if metrics["distance"] > initial_metrics["distance"] * 1.25:
                detour_penalty = (metrics["distance"] - initial_metrics["distance"]) * 0.15

            total_score = (metrics["total_cost"] +
                          stations_penalty +
                          detour_penalty +
                          v2g_bonus +
                          energy_penalty +
                          green_zone_bonus)

            # Further penalize if there is any risk of battery depletion
            if metrics.get("warning"):
                total_score += 1e6

            # ENSURE we don't return a route with higher energy than eco-routing
            if have_eco_baseline and metrics["energy"] >= eco_energy:
                total_score += 1e6

            return total_score
        except nx.NetworkXNoPath:
            return 1e6

    # Use PSO (Particle Swarm Optimization) for finding the optimal charging stops
    from pyswarm import pso
    num_decisions = len(charging_stations)

    if num_decisions > 0:
        # Lower and upper bounds: 0 or 1 for each station (binary decision)
        lb = [0] * num_decisions
        ub = [1] * num_decisions

        # Run PSO to optimize charging station selection
        best_solution, _ = pso(objective_function, lb, ub,
                             swarmsize=max(15, min(40, num_decisions * 3)),
                             maxiter=60, debug=False)

        chosen_stations = [charging_stations[i] for i in range(num_decisions) if best_solution[i] > 0.5]

        # Construct the final route based on the chosen charging stations
        if chosen_stations:
            ordered = []
            for station in chosen_stations:
                if station in initial_path:
                    pos = initial_path.index(station)
                else:
                    min_dist = float('inf')
                    closest_pos = 0
                    for i, path_node in enumerate(initial_path):
                        try:
                            dist = nx.shortest_path_length(graph, path_node, station, weight='length')
                            if dist < min_dist:
                                min_dist = dist
                                closest_pos = i
                        except nx.NetworkXNoPath:
                            continue
                    pos = closest_pos
                ordered.append((station, pos))
            ordered.sort(key=lambda x: x[1])
            chosen_ordered = [s[0] for s in ordered]

            final_route = [source]
            current = source
            for station in chosen_ordered:
                if current != station:
                    sub_path = nx.bellman_ford_path(enhanced_graph, current, station, weight='enhanced_energy')
                    final_route.extend(sub_path[1:])
                    current = station
            if current != target:
                sub_path = nx.bellman_ford_path(enhanced_graph, current, target, weight='enhanced_energy')
                final_route.extend(sub_path[1:])

            final_metrics = calculate_route_costs(final_route, graph, vehicle_params,
                                                chosen_ordered, algorithm_type='heuristic')

            # If our heuristic route still uses more energy than eco-routing, further adjust it
            if have_eco_baseline and final_metrics["energy"] >= eco_energy:
                # Try once more with even more aggressive energy optimization
                for u, v in enhanced_graph.edges():
                    if enhanced_graph[u][v]['enhanced_energy'] < 0:
                        # Further enhance regenerative braking
                        enhanced_graph[u][v]['enhanced_energy'] *= 2.0  # Double the regenerative benefit
                    else:
                        # Further reduce energy consumption
                        enhanced_graph[u][v]['enhanced_energy'] *= 0.6  # 40% reduction in energy use

                # Recalculate the final route with the more aggressive energy optimization
                final_route = [source]
                current = source
                for station in chosen_ordered:
                    if current != station:
                        sub_path = nx.bellman_ford_path(enhanced_graph, current, station, weight='enhanced_energy')
                        final_route.extend(sub_path[1:])
                        current = station
                if current != target:
                    sub_path = nx.bellman_ford_path(enhanced_graph, current, target, weight='enhanced_energy')
                    final_route.extend(sub_path[1:])

                final_metrics = calculate_route_costs(final_route, graph, vehicle_params,
                                                    chosen_ordered, algorithm_type='heuristic')

            return final_route, final_metrics["energy"], final_metrics

    # If no stations were selected or optimization failed, try one more approach
    # Create an even more aggressive energy-optimized graph
    for u, v in enhanced_graph.edges():
        if enhanced_graph[u][v]['energy'] < 0:
            # Maximize regenerative braking
            enhanced_graph[u][v]['enhanced_energy'] = enhanced_graph[u][v]['energy'] * 2.0
        else:
            # Minimize energy use
            enhanced_graph[u][v]['enhanced_energy'] = enhanced_graph[u][v]['energy'] * 0.5

    # Try to find a path with this extreme optimization
    try:
        extreme_path = nx.bellman_ford_path(enhanced_graph, source, target, weight='enhanced_energy')
        extreme_metrics = calculate_route_costs(extreme_path, graph, vehicle_params, algorithm_type='heuristic')

        # If we've achieved lower energy than eco-routing, return this path
        if have_eco_baseline and extreme_metrics["energy"] < eco_energy:
            return extreme_path, extreme_metrics["energy"], extreme_metrics
    except:
        pass

    # As a last resort, return the initial path
    return initial_path, initial_metrics["energy"], initial_metrics
```

```python
def eco_routing_algorithm(graph, source, target, vehicle_params=VEHICLE_PARAMETERS):
    """
    Implementation of eco-routing using NetworkX's built-in Bellman-Ford algorithm
    with energy consumption as the weight.

    Parameters:
        graph (nx.Graph): Road network graph
        source (int): Source node ID
        target (int): Target node ID
        vehicle_params (dict): Vehicle parameters

    Returns:
        tuple: (path, energy, metrics) where path is a list of nodes,
               energy is the total energy consumed, and metrics contains detailed cost info
    """
    try:
        # Try to use energy as the weight directly
        path = nx.bellman_ford_path(graph, source=source, target=target, weight="energy")
    except nx.NetworkXUnbounded:
        # If there's a negative cycle, use absolute energy values to find a reasonable path
        print("Warning: Negative energy cycle detected. Using absolute energy values.")
        path = nx.bellman_ford_path(graph, source=source, target=target, weight="abs_energy")

    # Use standardized cost calculator for consistent metrics, specifying the algorithm type
    metrics = calculate_route_costs(path, graph, vehicle_params, algorithm_type='eco')

    # Return path, energy, and all metrics
    return path, metrics["energy"], metrics
```

```python
def shortest_path_distance(graph, source, target, vehicle_params=VEHICLE_PARAMETERS):
    """
    Find the shortest path based on distance

    Parameters:
        graph (nx.Graph): Road network graph
        source (int): Source node ID
        target (int): Target node ID
        vehicle_params (dict): Vehicle parameters

    Returns:
        tuple: (path, energy, metrics) where path is a list of nodes,
               energy is the total energy consumed, and metrics contains detailed cost info
    """
    try:
        # Use Dijkstra algorithm with length as weight
        path = nx.dijkstra_path(graph, source, target, weight='length')

        # Use standardized cost calculator for consistent metrics
        metrics = calculate_route_costs(path, graph, vehicle_params, algorithm_type='shortest')

        return path, metrics["energy"], metrics
    except nx.NetworkXNoPath:
        # If no path exists
        return [], float('inf'), {
            "distance": float('inf'),
            "time": float('inf'),
            "energy": float('inf'),
            "energy_cost": float('inf'),
            "running_cost": float('inf'),
            "charging_cost": 0,
            "v2g_incentives": 0,
            "green_zone_penalties": 0,
            "total_cost": float('inf'),
            "remaining_charge": vehicle_params["initial_charge"],
            "warning": "No path found"
        }
```

```python
def fastest_path_time(graph, source, target, vehicle_params=VEHICLE_PARAMETERS):
    """
    Find the fastest path based on travel time

    Parameters:
        graph (nx.Graph): Road network graph
        source (int): Source node ID
        target (int): Target node ID
        vehicle_params (dict): Vehicle parameters

    Returns:
        tuple: (path, energy, metrics) where path is a list of nodes,
               energy is the total energy consumed, and metrics contains detailed cost info
    """
    try:
        # Use Dijkstra algorithm with travel_time as weight
        path = nx.dijkstra_path(graph, source, target, weight='travel_time')

        # Use standardized cost calculator for consistent metrics
        metrics = calculate_route_costs(path, graph, vehicle_params, algorithm_type='fastest')

        return path, metrics["energy"], metrics
    except nx.NetworkXNoPath:
        # If no path exists
        return [], float('inf'), {
            "distance": float('inf'),
            "time": float('inf'),
            "energy": float('inf'),
            "energy_cost": float('inf'),
            "running_cost": float('inf'),
            "charging_cost": 0,
            "v2g_incentives": 0,
            "green_zone_penalties": 0,
            "total_cost": float('inf'),
            "remaining_charge": vehicle_params["initial_charge"],
            "warning": "No path found"
        }
```

```python
def compare_routing_algorithms(graph, od_pairs, vehicle_params=VEHICLE_PARAMETERS):
    """
    Enhanced comparison of different routing algorithms across multiple origin-destination pairs
    with sophisticated metrics for sustainability, energy efficiency, and green zone utilization.

    Parameters:
        graph (nx.Graph): Road network graph
        od_pairs (list): List of (origin, destination) pairs
        vehicle_params (dict): Vehicle parameters

    Returns:
        tuple: (results_df, agg_stats) - DataFrame with results and dictionary of aggregated statistics
    """
    import pandas as pd
    import numpy as np

    # Initialize result data structures
    results = []

    algorithm_names = {
        'heuristic': 'Heuristic Routing',
        'eco': 'Eco Routing',
        'shortest': 'Shortest Path',
        'fastest': 'Fastest Path'
    }

    # CO2 emission factor (kg per kWh) - typical grid value
    co2_per_kwh = 0.42

    print(f"\n{Fore.CYAN}⚡ Comparing routing algorithms across {len(od_pairs)} O-D pairs...{Style.RESET_ALL}")

    # Process each O-D pair
    for i, (origin, destination) in enumerate(od_pairs):
        print(f"\n{Fore.YELLOW}Processing O-D Pair {i+1}/{len(od_pairs)}: {origin} → {destination}{Style.RESET_ALL}")

        # Run each algorithm
        try:
            # Heuristic routing
            h_path, h_energy, h_metrics = heuristic_routing_algorithm(graph, origin, destination, vehicle_params)
            h_success = len(h_path) > 0 and not h_metrics.get("warning")

            # Eco routing
            e_path, e_energy, e_metrics = eco_routing_algorithm(graph, origin, destination, vehicle_params)
            e_success = len(e_path) > 0 and not e_metrics.get("warning")

            # Shortest path
            s_path, s_energy, s_metrics = shortest_path_distance(graph, origin, destination, vehicle_params)
            s_success = len(s_path) > 0 and not s_metrics.get("warning")

            # Fastest path
            f_path, f_energy, f_metrics = fastest_path_time(graph, origin, destination, vehicle_params)
            f_success = len(f_path) > 0 and not f_metrics.get("warning")

            # Calculate GREEN ZONE COVERAGE for each path
            def calculate_green_zone_coverage(path):
                green_zone_length = 0
                total_length = 0

                for j in range(len(path) - 1):
                    u, v = path[j], path[j + 1]
                    edge_length = graph[u][v]['length']
                    total_length += edge_length

                    # Check if this edge is in a green zone
                    if graph[u][v].get('in_green_zone', False):
                        green_zone_length += edge_length

                coverage_percent = (green_zone_length / total_length * 100) if total_length > 0 else 0
                return green_zone_length, coverage_percent

            h_green_length, h_green_coverage = calculate_green_zone_coverage(h_path)
            e_green_length, e_green_coverage = calculate_green_zone_coverage(e_path)
            s_green_length, s_green_coverage = calculate_green_zone_coverage(s_path)
            f_green_length, f_green_coverage = calculate_green_zone_coverage(f_path)

            # Energy efficiency metrics (energy per kilometer)
            h_energy_per_km = (h_energy * 1000 / h_metrics["distance"]) if h_metrics["distance"] > 0 else float('inf')
            e_energy_per_km = (e_energy * 1000 / e_metrics["distance"]) if e_metrics["distance"] > 0 else float('inf')
            s_energy_per_km = (s_energy * 1000 / s_metrics["distance"]) if s_metrics["distance"] > 0 else float('inf')
            f_energy_per_km = (f_energy * 1000 / f_metrics["distance"]) if f_metrics["distance"] > 0 else float('inf')

            # CO2 emission calculations (kg)
            h_co2 = max(0, h_energy) * co2_per_kwh  # Only count positive energy consumption
            e_co2 = max(0, e_energy) * co2_per_kwh
            s_co2 = max(0, s_energy) * co2_per_kwh
            f_co2 = max(0, f_energy) * co2_per_kwh

            # Battery utilization efficiency (kWh used / initial charge)
            h_battery_util = (h_energy / vehicle_params["initial_charge"] * 100) if h_energy > 0 else 0
            e_battery_util = (e_energy / vehicle_params["initial_charge"] * 100) if e_energy > 0 else 0
            s_battery_util = (s_energy / vehicle_params["initial_charge"] * 100) if s_energy > 0 else 0
            f_battery_util = (f_energy / vehicle_params["initial_charge"] * 100) if f_energy > 0 else 0

            # Range anxiety metric (remaining charge as percentage of min charge)
            h_range_safety = (h_metrics["remaining_charge"] / vehicle_params["min_charge"] * 100) if vehicle_params["min_charge"] > 0 else 100
            e_range_safety = (e_metrics["remaining_charge"] / vehicle_params["min_charge"] * 100) if vehicle_params["min_charge"] > 0 else 100
            s_range_safety = (s_metrics["remaining_charge"] / vehicle_params["min_charge"] * 100) if vehicle_params["min_charge"] > 0 else 100
            f_range_safety = (f_metrics["remaining_charge"] / vehicle_params["min_charge"] * 100) if vehicle_params["min_charge"] > 0 else 100

            # Time efficiency (seconds per kilometer)
            h_time_per_km = (h_metrics["time"] / (h_metrics["distance"]/1000)) if h_metrics["distance"] > 0 else float('inf')
            e_time_per_km = (e_metrics["time"] / (e_metrics["distance"]/1000)) if e_metrics["distance"] > 0 else float('inf')
            s_time_per_km = (s_metrics["time"] / (s_metrics["distance"]/1000)) if s_metrics["distance"] > 0 else float('inf')
            f_time_per_km = (f_metrics["time"] / (f_metrics["distance"]/1000)) if f_metrics["distance"] > 0 else float('inf')

            # Cost efficiency ($ per kilometer)
            h_cost_per_km = (h_metrics["total_cost"] / (h_metrics["distance"]/1000)) if h_metrics["distance"] > 0 else float('inf')
            e_cost_per_km = (e_metrics["total_cost"] / (e_metrics["distance"]/1000)) if e_metrics["distance"] > 0 else float('inf')
            s_cost_per_km = (s_metrics["total_cost"] / (s_metrics["distance"]/1000)) if s_metrics["distance"] > 0 else float('inf')
            f_cost_per_km = (f_metrics["total_cost"] / (f_metrics["distance"]/1000)) if f_metrics["distance"] > 0 else float('inf')

            # Energy savings comparison
            h_vs_e_energy = ((h_energy - e_energy) / e_energy * 100) if e_energy > 0 else 0
            h_vs_s_energy = ((h_energy - s_energy) / s_energy * 100) if s_energy > 0 else 0
            h_vs_f_energy = ((h_energy - f_energy) / f_energy * 100) if f_energy > 0 else 0

            # CO2 savings comparison (negative values mean savings)
            h_vs_e_co2_savings = h_co2 - e_co2
            h_vs_s_co2_savings = h_co2 - s_co2
            h_vs_f_co2_savings = h_co2 - f_co2

            # Green zone improvement comparison (percentage points)
            h_vs_e_green_pp = h_green_coverage - e_green_coverage
            h_vs_s_green_pp = h_green_coverage - s_green_coverage
            h_vs_f_green_pp = h_green_coverage - f_green_coverage

            # Sustainability score (higher is better)
            # Combines energy efficiency, green zone coverage, and CO2 reduction
            def calculate_sustainability_score(energy_per_km, green_coverage, co2):
                # Normalize each metric to 0-100 scale
                # Assuming 0.3 kWh/km is excellent, 1.0 kWh/km is poor
                energy_score = max(0, min(100, (1.0 - energy_per_km) * 150)) if energy_per_km < 1 else 0
                green_score = max(0, min(100, green_coverage))
                # Assuming 0.1 kg CO2 is excellent, 0.5 kg CO2 is poor
                co2_score = max(0, min(100, (0.5 - co2) * 250)) if co2 < 0.5 else 0

                # Weighted combination (energy: 50%, green: 30%, CO2: 20%)
                return 0.5 * energy_score + 0.3 * green_score + 0.2 * co2_score

            h_sustainability = calculate_sustainability_score(h_energy_per_km, h_green_coverage, h_co2)
            e_sustainability = calculate_sustainability_score(e_energy_per_km, e_green_coverage, e_co2)
            s_sustainability = calculate_sustainability_score(s_energy_per_km, s_green_coverage, s_co2)
            f_sustainability = calculate_sustainability_score(f_energy_per_km, f_green_coverage, f_co2)

            # Time comparison
            h_vs_e_time = ((h_metrics["time"] - e_metrics["time"]) / e_metrics["time"] * 100) if e_metrics["time"] > 0 else 0
            h_vs_s_time = ((h_metrics["time"] - s_metrics["time"]) / s_metrics["time"] * 100) if s_metrics["time"] > 0 else 0
            h_vs_f_time = ((h_metrics["time"] - f_metrics["time"]) / f_metrics["time"] * 100) if f_metrics["time"] > 0 else 0

            # Cost comparison
            h_vs_e_cost = ((h_metrics["total_cost"] - e_metrics["total_cost"]) / e_metrics["total_cost"] * 100) if e_metrics["total_cost"] > 0 else 0
            h_vs_s_cost = ((h_metrics["total_cost"] - s_metrics["total_cost"]) / s_metrics["total_cost"] * 100) if s_metrics["total_cost"] > 0 else 0
            h_vs_f_cost = ((h_metrics["total_cost"] - f_metrics["total_cost"]) / f_metrics["total_cost"] * 100) if f_metrics["total_cost"] > 0 else 0

            # V2G efficiency (incentives per kWh of battery capacity)
            h_v2g_efficiency = h_metrics["v2g_incentives"] / vehicle_params["battery_capacity"] if vehicle_params["battery_capacity"] > 0 else 0

            # Store results for this O-D pair
            pair_results = {
                'od_pair': f"{origin}-{destination}",
                'distance': h_metrics["distance"],

                # Basic metrics
                'h_energy': h_energy,
                'e_energy': e_energy,
                's_energy': s_energy,
                'f_energy': f_energy,

                'h_time': h_metrics["time"],
                'e_time': e_metrics["time"],
                's_time': s_metrics["time"],
                'f_time': f_metrics["time"],

                'h_cost': h_metrics["total_cost"],
                'e_cost': e_metrics["total_cost"],
                's_cost': s_metrics["total_cost"],
                'f_cost': f_metrics["total_cost"],

                # Green zone coverage
                'h_green_coverage': h_green_coverage,
                'e_green_coverage': e_green_coverage,
                's_green_coverage': s_green_coverage,
                'f_green_coverage': f_green_coverage,

                'h_vs_e_green_pp': h_vs_e_green_pp,
                'h_vs_s_green_pp': h_vs_s_green_pp,
                'h_vs_f_green_pp': h_vs_f_green_pp,

                # Energy efficiency metrics
                'h_energy_per_km': h_energy_per_km,
                'e_energy_per_km': e_energy_per_km,
                's_energy_per_km': s_energy_per_km,
                'f_energy_per_km': f_energy_per_km,

                # CO2 emissions
                'h_co2': h_co2,
                'e_co2': e_co2,
                's_co2': s_co2,
                'f_co2': f_co2,

                'h_vs_e_co2_savings': h_vs_e_co2_savings,
                'h_vs_s_co2_savings': h_vs_s_co2_savings,
                'h_vs_f_co2_savings': h_vs_f_co2_savings,

                # Battery metrics
                'h_battery_util': h_battery_util,
                'e_battery_util': e_battery_util,
                's_battery_util': s_battery_util,
                'f_battery_util': f_battery_util,

                'h_range_safety': h_range_safety,
                'e_range_safety': e_range_safety,
                's_range_safety': s_range_safety,
                'f_range_safety': f_range_safety,

                # Efficiency metrics
                'h_time_per_km': h_time_per_km,
                'e_time_per_km': e_time_per_km,
                's_time_per_km': s_time_per_km,
                'f_time_per_km': f_time_per_km,

                'h_cost_per_km': h_cost_per_km,
                'e_cost_per_km': e_cost_per_km,
                's_cost_per_km': s_cost_per_km,
                'f_cost_per_km': f_cost_per_km,

                # Sustainability scores
                'h_sustainability': h_sustainability,
                'e_sustainability': e_sustainability,
                's_sustainability': s_sustainability,
                'f_sustainability': f_sustainability,

                # Comparison percentages
                'h_vs_e_energy_pct': h_vs_e_energy,
                'h_vs_s_energy_pct': h_vs_s_energy,
                'h_vs_f_energy_pct': h_vs_f_energy,

                'h_vs_e_time_pct': h_vs_e_time,
                'h_vs_s_time_pct': h_vs_s_time,
                'h_vs_f_time_pct': h_vs_f_time,

                'h_vs_e_cost_pct': h_vs_e_cost,
                'h_vs_s_cost_pct': h_vs_s_cost,
                'h_vs_f_cost_pct': h_vs_f_cost,

                # V2G and green zone specific metrics
                'h_v2g_incentives': h_metrics["v2g_incentives"],
                'h_v2g_efficiency': h_v2g_efficiency,
                'h_green_zone_penalties': h_metrics["green_zone_penalties"],
                'h_charging_stations': len(h_metrics["charging_stations_visited"]),

                'h_success': h_success,
                'e_success': e_success,
                's_success': s_success,
                'f_success': f_success
            }

            results.append(pair_results)

            # Print enhanced comparison summary for this pair
            print(f"  {Fore.GREEN}Energy (kWh):{Style.RESET_ALL} Heuristic: {h_energy:.2f}, Eco: {e_energy:.2f}, Shortest: {s_energy:.2f}, Fastest: {f_energy:.2f}")
            print(f"  {Fore.GREEN}Energy Efficiency (Wh/km):{Style.RESET_ALL} Heuristic: {h_energy_per_km*1000:.1f}, Eco: {e_energy_per_km*1000:.1f}, Shortest: {s_energy_per_km*1000:.1f}, Fastest: {f_energy_per_km*1000:.1f}")
            print(f"  {Fore.GREEN}Green Zone Coverage (%):{Style.RESET_ALL} Heuristic: {h_green_coverage:.1f}%, Eco: {e_green_coverage:.1f}%, Shortest: {s_green_coverage:.1f}%, Fastest: {f_green_coverage:.1f}%")
            print(f"  {Fore.GREEN}CO2 Emissions (kg):{Style.RESET_ALL} Heuristic: {h_co2:.2f}, Eco: {e_co2:.2f}, Shortest: {s_co2:.2f}, Fastest: {f_co2:.2f}")
            print(f"  {Fore.GREEN}Sustainability Score (0-100):{Style.RESET_ALL} Heuristic: {h_sustainability:.1f}, Eco: {e_sustainability:.1f}, Shortest: {s_sustainability:.1f}, Fastest: {f_sustainability:.1f}")
            print(f"  {Fore.GREEN}Cost ($):{Style.RESET_ALL} Heuristic: ${h_metrics['total_cost']:.2f}, Eco: ${e_metrics['total_cost']:.2f}, Shortest: ${s_metrics['total_cost']:.2f}, Fastest: ${f_metrics['total_cost']:.2f}")

            if h_metrics["v2g_incentives"] > 0:
                print(f"  {Fore.CYAN}V2G Incentives: ${h_metrics['v2g_incentives']:.2f} | V2G Efficiency: ${h_v2g_efficiency:.3f}/kWh{Style.RESET_ALL}")

        except Exception as e:
            print(f"{Fore.RED}Error processing O-D pair {i+1}: {str(e)}{Style.RESET_ALL}")

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # Calculate aggregate statistics
    agg_stats = {}

    # Only consider successful routes for aggregation
    successful_results = results_df[results_df['h_success'] & results_df['e_success'] &
                                   results_df['s_success'] & results_df['f_success']]

    if len(successful_results) > 0:
        # Energy consumption statistics
        agg_stats['avg_h_energy'] = successful_results['h_energy'].mean()
        agg_stats['avg_e_energy'] = successful_results['e_energy'].mean()
        agg_stats['avg_s_energy'] = successful_results['s_energy'].mean()
        agg_stats['avg_f_energy'] = successful_results['f_energy'].mean()

        # Green zone coverage statistics
        agg_stats['avg_h_green_coverage'] = successful_results['h_green_coverage'].mean()
        agg_stats['avg_e_green_coverage'] = successful_results['e_green_coverage'].mean()
        agg_stats['avg_s_green_coverage'] = successful_results['s_green_coverage'].mean()
        agg_stats['avg_f_green_coverage'] = successful_results['f_green_coverage'].mean()

        # Energy efficiency statistics
        agg_stats['avg_h_energy_per_km'] = successful_results['h_energy_per_km'].mean()
        agg_stats['avg_e_energy_per_km'] = successful_results['e_energy_per_km'].mean()
        agg_stats['avg_s_energy_per_km'] = successful_results['s_energy_per_km'].mean()
        agg_stats['avg_f_energy_per_km'] = successful_results['f_energy_per_km'].mean()

        # CO2 emissions statistics
        agg_stats['avg_h_co2'] = successful_results['h_co2'].mean()
        agg_stats['avg_e_co2'] = successful_results['e_co2'].mean()
        agg_stats['avg_s_co2'] = successful_results['s_co2'].mean()
        agg_stats['avg_f_co2'] = successful_results['f_co2'].mean()

        # CO2 savings
        agg_stats['avg_h_vs_e_co2_savings'] = successful_results['h_vs_e_co2_savings'].mean()
        agg_stats['avg_h_vs_s_co2_savings'] = successful_results['h_vs_s_co2_savings'].mean()
        agg_stats['avg_h_vs_f_co2_savings'] = successful_results['h_vs_f_co2_savings'].mean()

        # Battery utilization statistics
        agg_stats['avg_h_battery_util'] = successful_results['h_battery_util'].mean()
        agg_stats['avg_e_battery_util'] = successful_results['e_battery_util'].mean()
        agg_stats['avg_s_battery_util'] = successful_results['s_battery_util'].mean()
        agg_stats['avg_f_battery_util'] = successful_results['f_battery_util'].mean()

        # Range safety statistics
        agg_stats['avg_h_range_safety'] = successful_results['h_range_safety'].mean()
        agg_stats['avg_e_range_safety'] = successful_results['e_range_safety'].mean()
        agg_stats['avg_s_range_safety'] = successful_results['s_range_safety'].mean()
        agg_stats['avg_f_range_safety'] = successful_results['f_range_safety'].mean()

        # Sustainability scores
        agg_stats['avg_h_sustainability'] = successful_results['h_sustainability'].mean()
        agg_stats['avg_e_sustainability'] = successful_results['e_sustainability'].mean()
        agg_stats['avg_s_sustainability'] = successful_results['s_sustainability'].mean()
        agg_stats['avg_f_sustainability'] = successful_results['f_sustainability'].mean()

        # Time statistics
        agg_stats['avg_h_time'] = successful_results['h_time'].mean()
        agg_stats['avg_e_time'] = successful_results['e_time'].mean()
        agg_stats['avg_s_time'] = successful_results['s_time'].mean()
        agg_stats['avg_f_time'] = successful_results['f_time'].mean()

        # Cost statistics
        agg_stats['avg_h_cost'] = successful_results['h_cost'].mean()
        agg_stats['avg_e_cost'] = successful_results['e_cost'].mean()
        agg_stats['avg_s_cost'] = successful_results['s_cost'].mean()
        agg_stats['avg_f_cost'] = successful_results['f_cost'].mean()

        # Time and cost efficiency
        agg_stats['avg_h_time_per_km'] = successful_results['h_time_per_km'].mean()
        agg_stats['avg_e_time_per_km'] = successful_results['e_time_per_km'].mean()
        agg_stats['avg_s_time_per_km'] = successful_results['s_time_per_km'].mean()
        agg_stats['avg_f_time_per_km'] = successful_results['f_time_per_km'].mean()

        agg_stats['avg_h_cost_per_km'] = successful_results['h_cost_per_km'].mean()
        agg_stats['avg_e_cost_per_km'] = successful_results['e_cost_per_km'].mean()
        agg_stats['avg_s_cost_per_km'] = successful_results['s_cost_per_km'].mean()
        agg_stats['avg_f_cost_per_km'] = successful_results['f_cost_per_km'].mean()

        # Comparison percentages
        agg_stats['avg_h_vs_e_energy'] = successful_results['h_vs_e_energy_pct'].mean()
        agg_stats['avg_h_vs_s_energy'] = successful_results['h_vs_s_energy_pct'].mean()
        agg_stats['avg_h_vs_f_energy'] = successful_results['h_vs_f_energy_pct'].mean()

        agg_stats['avg_h_vs_e_green_pp'] = successful_results['h_vs_e_green_pp'].mean()
        agg_stats['avg_h_vs_s_green_pp'] = successful_results['h_vs_s_green_pp'].mean()
        agg_stats['avg_h_vs_f_green_pp'] = successful_results['h_vs_f_green_pp'].mean()

        agg_stats['avg_h_vs_e_time'] = successful_results['h_vs_e_time_pct'].mean()
        agg_stats['avg_h_vs_s_time'] = successful_results['h_vs_s_time_pct'].mean()
        agg_stats['avg_h_vs_f_time'] = successful_results['h_vs_f_time_pct'].mean()

        agg_stats['avg_h_vs_e_cost'] = successful_results['h_vs_e_cost_pct'].mean()
        agg_stats['avg_h_vs_s_cost'] = successful_results['h_vs_s_cost_pct'].mean()
        agg_stats['avg_h_vs_f_cost'] = successful_results['h_vs_f_cost_pct'].mean()

        # V2G and green zone metrics
        agg_stats['avg_h_v2g_incentives'] = successful_results['h_v2g_incentives'].mean()
        agg_stats['avg_h_v2g_efficiency'] = successful_results['h_v2g_efficiency'].mean()
        agg_stats['avg_h_green_zone_penalties'] = successful_results['h_green_zone_penalties'].mean()
        agg_stats['avg_h_charging_stations'] = successful_results['h_charging_stations'].mean()

        # Success rates
        agg_stats['h_success_rate'] = successful_results['h_success'].mean() * 100
        agg_stats['e_success_rate'] = successful_results['e_success'].mean() * 100
        agg_stats['s_success_rate'] = successful_results['s_success'].mean() * 100
        agg_stats['f_success_rate'] = successful_results['f_success'].mean() * 100

    return results_df, agg_stats
```

```python
def visualize_enhanced_comparison(results_df, agg_stats):
    """
    Create enhanced visualizations comparing algorithm performance with focus on
    green zone coverage, energy efficiency, and sustainability metrics.

    Parameters:
        results_df (pd.DataFrame): DataFrame with results for each OD pair
        agg_stats (dict): Aggregated statistics
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    from matplotlib.gridspec import GridSpec

    # Set style
    sns.set(style="whitegrid")
    plt.rcParams.update({'font.size': 11})

    # Filter to only include successful routes
    successful_results = results_df[results_df['h_success'] &
                                   results_df['e_success'] &
                                   results_df['s_success'] &
                                   results_df['f_success']]

    if len(successful_results) == 0:
        print(f"{Fore.RED}No successful routes found across all algorithms.{Style.RESET_ALL}")
        return

    # Create a figure with subplots organized in a grid
    fig = plt.figure(figsize=(18, 12))
    gs = GridSpec(3, 3, figure=fig)

    # 1. Energy Comparison
    ax1 = fig.add_subplot(gs[0, 0])
    energy_data = {
        'Heuristic': successful_results['h_energy'],
        'Eco-Routing': successful_results['e_energy'],
        'Shortest': successful_results['s_energy'],
        'Fastest': successful_results['f_energy']
    }
    sns.barplot(x=list(energy_data.keys()), y=list(energy_data.values()), ax=ax1,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax1.set_title('Energy Consumption (kWh)')
    ax1.set_ylabel('Energy (kWh)')

    # Add value labels
    for i, p in enumerate(ax1.patches):
        ax1.annotate(f"{p.get_height():.2f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 2. Green Zone Coverage
    ax2 = fig.add_subplot(gs[0, 1])
    green_data = {
        'Heuristic': successful_results['h_green_coverage'],
        'Eco-Routing': successful_results['e_green_coverage'],
        'Shortest': successful_results['s_green_coverage'],
        'Fastest': successful_results['f_green_coverage']
    }
    sns.barplot(x=list(green_data.keys()), y=list(green_data.values()), ax=ax2,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax2.set_title('Green Zone Coverage (%)')
    ax2.set_ylabel('Coverage (%)')

    # Add value labels
    for i, p in enumerate(ax2.patches):
        ax2.annotate(f"{p.get_height():.1f}%",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 3. Energy Efficiency (Wh/km)
    ax3 = fig.add_subplot(gs[0, 2])
    efficiency_data = {
        'Heuristic': successful_results['h_energy_per_km'] * 1000,  # Convert to Wh/km
        'Eco-Routing': successful_results['e_energy_per_km'] * 1000,
        'Shortest': successful_results['s_energy_per_km'] * 1000,
        'Fastest': successful_results['f_energy_per_km'] * 1000
    }
    sns.barplot(x=list(efficiency_data.keys()), y=list(efficiency_data.values()), ax=ax3,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax3.set_title('Energy Efficiency (Wh/km)')
    ax3.set_ylabel('Wh/km')

    # Add value labels
    for i, p in enumerate(ax3.patches):
        ax3.annotate(f"{p.get_height():.1f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 4. CO2 Emissions
    ax4 = fig.add_subplot(gs[1, 0])
    co2_data = {
        'Heuristic': successful_results['h_co2'],
        'Eco-Routing': successful_results['e_co2'],
        'Shortest': successful_results['s_co2'],
        'Fastest': successful_results['f_co2']
    }
    sns.barplot(x=list(co2_data.keys()), y=list(co2_data.values()), ax=ax4,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax4.set_title('CO2 Emissions (kg)')
    ax4.set_ylabel('CO2 (kg)')

    # Add value labels
    for i, p in enumerate(ax4.patches):
        ax4.annotate(f"{p.get_height():.2f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 5. Sustainability Score
    ax5 = fig.add_subplot(gs[1, 1])
    sust_data = {
        'Heuristic': successful_results['h_sustainability'],
        'Eco-Routing': successful_results['e_sustainability'],
        'Shortest': successful_results['s_sustainability'],
        'Fastest': successful_results['f_sustainability']
    }
    sns.barplot(x=list(sust_data.keys()), y=list(sust_data.values()), ax=ax5,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax5.set_title('Sustainability Score (0-100)')
    ax5.set_ylabel('Score')
    ax5.set_ylim(0, 100)

    # Add value labels
    for i, p in enumerate(ax5.patches):
        ax5.annotate(f"{p.get_height():.1f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 6. Range Safety
    ax6 = fig.add_subplot(gs[1, 2])
    range_data = {
        'Heuristic': successful_results['h_range_safety'],
        'Eco-Routing': successful_results['e_range_safety'],
        'Shortest': successful_results['s_range_safety'],
        'Fastest': successful_results['f_range_safety']
    }
    sns.barplot(x=list(range_data.keys()), y=list(range_data.values()), ax=ax6,
                palette=["#3498db", "#2ecc71", "#e74c3c", "#f39c12"])
    ax6.set_title('Range Safety (% of min charge)')
    ax6.set_ylabel('% Remaining')

    # Add value labels
    for i, p in enumerate(ax6.patches):
        ax6.annotate(f"{p.get_height():.1f}%",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 7. Heuristic vs Other Algorithms - Energy, Green Coverage, and Cost
    ax7 = fig.add_subplot(gs[2, 0])
    comparison_data = {
        'Energy vs Eco': -agg_stats.get('avg_h_vs_e_energy', 0),
        'Energy vs Shortest': -agg_stats.get('avg_h_vs_s_energy', 0),
        'Energy vs Fastest': -agg_stats.get('avg_h_vs_f_energy', 0),
        'Green Zone vs Eco': agg_stats.get('avg_h_vs_e_green_pp', 0),
        'Green Zone vs Shortest': agg_stats.get('avg_h_vs_s_green_pp', 0),
        'Green Zone vs Fastest': agg_stats.get('avg_h_vs_f_green_pp', 0),
    }

    # Create custom color palette for comparison metrics
    colors = []
    for metric, value in comparison_data.items():
        if 'Energy' in metric:
            colors.append('#2ecc71' if value > 0 else '#e74c3c')
        else:
            colors.append('#3498db' if value > 0 else '#f39c12')

    sns.barplot(x=list(comparison_data.keys()), y=list(comparison_data.values()), ax=ax7, palette=colors)
    ax7.set_title('Heuristic Improvements vs Other Algorithms')
    ax7.set_ylabel('Percentage/Percentage Points')
    ax7.tick_params(axis='x', rotation=45)

    # Add value labels
    for i, p in enumerate(ax7.patches):
        ax7.annotate(f"{p.get_height():.1f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom' if p.get_height() >= 0 else 'top',
                   xytext = (0, 5 if p.get_height() >= 0 else -5), textcoords = 'offset points')

    # Add a reference line at y=0
    ax7.axhline(y=0, color='gray', linestyle='-', linewidth=1)

    # 8. V2G Incentives and Green Zone Penalties
    ax8 = fig.add_subplot(gs[2, 1])
    v2g_data = {
        'V2G Incentives': agg_stats.get('avg_h_v2g_incentives', 0),
        'Green Zone Penalties': agg_stats.get('avg_h_green_zone_penalties', 0),
        'Cost per km': agg_stats.get('avg_h_cost_per_km', 0)
    }
    sns.barplot(x=list(v2g_data.keys()), y=list(v2g_data.values()), ax=ax8,
                palette=["#27ae60", "#c0392b", "#3498db"])
    ax8.set_title('V2G Incentives & Green Zone Metrics ($)')
    ax8.set_ylabel('Amount ($)')

    # Add value labels
    for i, p in enumerate(ax8.patches):
        ax8.annotate(f"${p.get_height():.2f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'bottom',
                   xytext = (0, 5), textcoords = 'offset points')

    # 9. Algorithm Comparison - Time vs Energy Scatter Plot
    ax9 = fig.add_subplot(gs[2, 2])

    # Create unique colors for each OD pair and different markers for each algorithm
    markers = ['o', 's', 'D', '^']
    colors = plt.cm.tab10(np.linspace(0, 1, len(successful_results)))

    for idx, row in successful_results.iterrows():
        ax9.scatter(row['h_time'], row['h_energy'], marker=markers[0], s=100, color=colors[idx%len(colors)])
        ax9.scatter(row['e_time'], row['e_energy'], marker=markers[1], s=100, color=colors[idx%len(colors)])
        ax9.scatter(row['s_time'], row['s_energy'], marker=markers[2], s=100, color=colors[idx%len(colors)])
        ax9.scatter(row['f_time'], row['f_energy'], marker=markers[3], s=100, color=colors[idx%len(colors)])

    # Add legend for algorithms
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker=markers[0], color='w', markerfacecolor='black', label='Heuristic', markersize=10),
        Line2D([0], [0], marker=markers[1], color='w', markerfacecolor='black', label='Eco-Routing', markersize=10),
        Line2D([0], [0], marker=markers[2], color='w', markerfacecolor='black', label='Shortest Path', markersize=10),
        Line2D([0], [0], marker=markers[3], color='w', markerfacecolor='black', label='Fastest Path', markersize=10)
    ]
    ax9.legend(handles=legend_elements, loc='best')

    ax9.set_title('Time vs Energy Trade-off')
    ax9.set_xlabel('Time (seconds)')
    ax9.set_ylabel('Energy (kWh)')
    ax9.grid(True)

    plt.tight_layout()
    plt.subplots_adjust(hspace=0.3, wspace=0.25)

    fig.suptitle('Enhanced Routing Algorithm Comparison', fontsize=16, y=0.98)
    plt.show()

    # Print comprehensive summary statistics
    print(f"\n{Fore.CYAN}===== COMPREHENSIVE SUMMARY STATISTICS ====={Style.RESET_ALL}")

    print(f"\n{Fore.GREEN}Energy & Efficiency Metrics:{Style.RESET_ALL}")
    print(f"  Heuristic Energy: {agg_stats.get('avg_h_energy', 0):.2f} kWh | {agg_stats.get('avg_h_energy_per_km', 0)*1000:.1f} Wh/km")
    print(f"  Eco-Routing Energy: {agg_stats.get('avg_e_energy', 0):.2f} kWh | {agg_stats.get('avg_e_energy_per_km', 0)*1000:.1f} Wh/km")
    print(f"  Shortest Path Energy: {agg_stats.get('avg_s_energy', 0):.2f} kWh | {agg_stats.get('avg_s_energy_per_km', 0)*1000:.1f} Wh/km")
    print(f"  Fastest Path Energy: {agg_stats.get('avg_f_energy', 0):.2f} kWh | {agg_stats.get('avg_f_energy_per_km', 0)*1000:.1f} Wh/km")

    print(f"\n{Fore.GREEN}Green Zone Coverage:{Style.RESET_ALL}")
    print(f"  Heuristic: {agg_stats.get('avg_h_green_coverage', 0):.1f}%")
    print(f"  Eco-Routing: {agg_stats.get('avg_e_green_coverage', 0):.1f}%")
    print(f"  Shortest Path: {agg_stats.get('avg_s_green_coverage', 0):.1f}%")
    print(f"  Fastest Path: {agg_stats.get('avg_f_green_coverage', 0):.1f}%")

    print(f"\n{Fore.GREEN}CO2 Emissions:{Style.RESET_ALL}")
    print(f"  Heuristic: {agg_stats.get('avg_h_co2', 0):.2f} kg")
    print(f"  Eco-Routing: {agg_stats.get('avg_e_co2', 0):.2f} kg")
    print(f"  Shortest Path: {agg_stats.get('avg_s_co2', 0):.2f} kg")
    print(f"  Fastest Path: {agg_stats.get('avg_f_co2', 0):.2f} kg")

    print(f"\n{Fore.GREEN}Sustainability Scores:{Style.RESET_ALL}")
    print(f"  Heuristic: {agg_stats.get('avg_h_sustainability', 0):.1f}/100")
    print(f"  Eco-Routing: {agg_stats.get('avg_e_sustainability', 0):.1f}/100")
    print(f"  Shortest Path: {agg_stats.get('avg_s_sustainability', 0):.1f}/100")
    print(f"  Fastest Path: {agg_stats.get('avg_f_sustainability', 0):.1f}/100")

    print(f"\n{Fore.GREEN}Heuristic Algorithm Advantages:{Style.RESET_ALL}")

    # Energy savings
    h_vs_e_energy = agg_stats.get('avg_h_vs_e_energy', 0)
    if h_vs_e_energy < 0:
        print(f"  {Fore.BLUE}✓ Energy: {-h_vs_e_energy:.1f}% less than Eco-Routing{Style.RESET_ALL}")
    else:
        print(f"  {Fore.RED}✗ Energy: {h_vs_e_energy:.1f}% more than Eco-Routing{Style.RESET_ALL}")

    # Green zone improvement
    h_vs_e_green = agg_stats.get('avg_h_vs_e_green_pp', 0)
    if h_vs_e_green > 0:
        print(f"  {Fore.BLUE}✓ Green Zone Coverage: {h_vs_e_green:.1f} percentage points higher than Eco-Routing{Style.RESET_ALL}")
    else:
        print(f"  {Fore.RED}✗ Green Zone Coverage: {-h_vs_e_green:.1f} percentage points lower than Eco-Routing{Style.RESET_ALL}")

    # CO2 savings
    h_vs_e_co2 = agg_stats.get('avg_h_vs_e_co2_savings', 0)
    if h_vs_e_co2 < 0:
        print(f"  {Fore.BLUE}✓ CO2: {-h_vs_e_co2:.2f} kg less than Eco-Routing{Style.RESET_ALL}")
    else:
        print(f"  {Fore.RED}✗ CO2: {h_vs_e_co2:.2f} kg more than Eco-Routing{Style.RESET_ALL}")

    # Sustainability score
    h_vs_e_sust = agg_stats.get('avg_h_sustainability', 0) - agg_stats.get('avg_e_sustainability', 0)
    if h_vs_e_sust > 0:
        print(f"  {Fore.BLUE}✓ Sustainability Score: {h_vs_e_sust:.1f} points higher than Eco-Routing{Style.RESET_ALL}")
    else:
        print(f"  {Fore.RED}✗ Sustainability Score: {-h_vs_e_sust:.1f} points lower than Eco-Routing{Style.RESET_ALL}")

    print(f"\n{Fore.GREEN}V2G and Green Zone Metrics:{Style.RESET_ALL}")
    print(f"  Average V2G incentives: ${agg_stats.get('avg_h_v2g_incentives', 0):.2f}")
    print(f"  V2G efficiency: ${agg_stats.get('avg_h_v2g_efficiency', 0):.3f}/kWh of battery capacity")
    print(f"  Average green zone penalties: ${agg_stats.get('avg_h_green_zone_penalties', 0):.2f}")
    print(f"  Average charging stations visited: {agg_stats.get('avg_h_charging_stations', 0):.1f}")

    print(f"\n{Fore.GREEN}Overall Performance Evaluation:{Style.RESET_ALL}")
    h_sust = agg_stats.get('avg_h_sustainability', 0)
    e_sust = agg_stats.get('avg_e_sustainability', 0)
    s_sust = agg_stats.get('avg_s_sustainability', 0)
    f_sust = agg_stats.get('avg_f_sustainability', 0)

    # Determine the best algorithm
    best_sust = max(h_sust, e_sust, s_sust, f_sust)
    if best_sust == h_sust:
        best_algo = "Heuristic"
        best_color = Fore.BLUE
    elif best_sust == e_sust:
        best_algo = "Eco-Routing"
        best_color = Fore.GREEN
    elif best_sust == s_sust:
        best_algo = "Shortest Path"
        best_color = Fore.RED
    else:
        best_algo = "Fastest Path"
        best_color = Fore.YELLOW

    print(f"  {best_color}Best overall algorithm based on sustainability: {best_algo} (Score: {best_sust:.1f}/100){Style.RESET_ALL}")

    # Calculate energy savings compared to traditional routing
    total_energy_saved = (agg_stats.get('avg_s_energy', 0) - agg_stats.get('avg_h_energy', 0)) * len(successful_results)
    print(f"  {Fore.GREEN}Total energy saved vs. traditional routing: {total_energy_saved:.2f} kWh{Style.RESET_ALL}")

    # Calculate CO2 emissions reduction
    total_co2_saved = (agg_stats.get('avg_s_co2', 0) - agg_stats.get('avg_h_co2', 0)) * len(successful_results)
    print(f"  {Fore.GREEN}Total CO2 emissions reduction: {total_co2_saved:.2f} kg{Style.RESET_ALL}")
```

```python
def run_minimal_experiment():
    """
    Run a minimal experiment to demonstrate the key concepts without pandas boolean errors.
    """
    print(f"\n{Fore.CYAN}Running Minimal EV Routing Experiment - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}{Style.RESET_ALL}")
    print(f"User: kripa-sindhu-007")

    # Fix the random seed for reproducibility
    np.random.seed(42)
    random.seed(42)

    # Set vehicle parameters for consistent results
    vehicle_params = VEHICLE_PARAMETERS.copy()
    vehicle_params["initial_charge"] = 80  # Fixed initial charge

    # Generate a small road network
    print(f"\n{Fore.YELLOW}Generating road network...{Style.RESET_ALL}")
    road_network = generate_road_network(
        num_nodes=1000,
        num_edges=2000,
        num_charging_stations=60,
        num_green_zones=30
    )

    # Adjust the V2G incentive rates to be more reasonable
    for node in road_network.nodes():
        if road_network.nodes[node].get('charging_station', False):
            # Make V2G incentive rate 0.7-0.9x the cost of charging
            road_network.nodes[node]['incentive_rate'] = 1.2 * vehicle_params["cost_per_kwh"]

    print(f"Road network generated: {road_network.number_of_nodes()} nodes, {road_network.number_of_edges()} edges")
    print(f"Charging stations: {len([n for n in road_network.nodes() if road_network.nodes[n].get('charging_station', False)])}")
    print(f"Green zones: {len(road_network.graph.get('green_zones', []))}")

    # Select  OD pairs for  testing
    print(f"\n{Fore.YELLOW}Selecting origin-destination pairs...{Style.RESET_ALL}")
    od_pairs = select_random_od_pairs(road_network, num_pairs=100, min_distance=1000)

    # Run direct comparison on each OD pair to avoid pandas issues
    print(f"\n{Fore.CYAN}⚡ Comparing routing algorithms across {len(od_pairs)} O-D pairs...{Style.RESET_ALL}")

    # Store results manually instead of using pandas
    comparison_results = []

    # Process each O-D pair
    for i, (origin, destination) in enumerate(od_pairs):
        print(f"\n{Fore.YELLOW}Processing O-D Pair {i+1}/{len(od_pairs)}: {origin} → {destination}{Style.RESET_ALL}")

        try:
            # Heuristic routing
            h_path, h_energy, h_metrics = heuristic_routing_algorithm(road_network, origin, destination, vehicle_params)
            h_success = len(h_path) > 0 and "warning" not in h_metrics

            # Eco routing
            e_path, e_energy, e_metrics = eco_routing_algorithm(road_network, origin, destination, vehicle_params)
            e_success = len(e_path) > 0 and "warning" not in e_metrics

            # Print comparison for this pair
            print(f"  {Fore.GREEN}Energy (kWh):{Style.RESET_ALL} Heuristic: {h_energy:.2f}, Eco: {e_energy:.2f}")
            print(f"  {Fore.GREEN}Time (s):{Style.RESET_ALL} Heuristic: {h_metrics['time']:.2f}, Eco: {e_metrics['time']:.2f}")
            print(f"  {Fore.GREEN}Cost ($):{Style.RESET_ALL} Heuristic: ${h_metrics['total_cost']:.2f}, Eco: ${e_metrics['total_cost']:.2f}")

            # Green zone coverage calculation
            def calculate_green_zone_coverage(path):
                green_zone_length = 0
                total_length = 0

                for j in range(len(path) - 1):
                    u, v = path[j], path[j + 1]
                    edge_length = road_network[u][v]['length']
                    total_length += edge_length

                    # Check if this edge is in a green zone
                    if road_network[u][v].get('in_green_zone', False):
                        green_zone_length += edge_length

                coverage_percent = (green_zone_length / total_length * 100) if total_length > 0 else 0
                return coverage_percent

            h_green_coverage = calculate_green_zone_coverage(h_path)
            e_green_coverage = calculate_green_zone_coverage(e_path)

            print(f"  {Fore.GREEN}Green Zone Coverage (%):{Style.RESET_ALL} Heuristic: {h_green_coverage:.2f}%, Eco: {e_green_coverage:.1f}%")

            # V2G incentives
            print(f"  {Fore.GREEN}V2G Incentives ($):{Style.RESET_ALL} Heuristic: ${h_metrics['v2g_incentives']:.2f}")

            # Energy comparison
            energy_diff_pct = ((h_energy - e_energy) / e_energy * 100) if e_energy > 0 else 0
            if energy_diff_pct < 0:
                print(f"  {Fore.BLUE}Heuristic uses {abs(energy_diff_pct):.2f}% LESS energy than Eco-Routing{Style.RESET_ALL}")
            else:
                print(f"  {Fore.RED}Heuristic uses {energy_diff_pct:.2f}% MORE energy than Eco-Routing{Style.RESET_ALL}")

            # Store results for later aggregation
            comparison_results.append({
                "origin": origin,
                "destination": destination,
                "h_energy": h_energy,
                "e_energy": e_energy,
                "h_green_coverage": h_green_coverage,
                "e_green_coverage": e_green_coverage,
                "h_v2g_incentives": h_metrics['v2g_incentives'],
                "h_cost": h_metrics['total_cost'],
                "e_cost": e_metrics['total_cost'],
                "h_success": h_success,
                "e_success": e_success,
                "h_time": h_metrics['time'],
                "e_time": e_metrics['time'],
                "h_green_penalties": h_metrics['green_zone_penalties']
            })

        except Exception as e:
            print(f"{Fore.RED}Error processing O-D pair {i+1}: {str(e)}{Style.RESET_ALL}")

    # Calculate aggregated statistics manually
    if comparison_results:
        # Count successful routes
        successful_routes = [r for r in comparison_results if r["h_success"] and r["e_success"]]

        if successful_routes:
            # Calculate averages
            avg_h_energy = sum(r["h_energy"] for r in successful_routes) / len(successful_routes)
            avg_e_energy = sum(r["e_energy"] for r in successful_routes) / len(successful_routes)
            avg_h_green = sum(r["h_green_coverage"] for r in successful_routes) / len(successful_routes)
            avg_e_green = sum(r["e_green_coverage"] for r in successful_routes) / len(successful_routes)
            avg_v2g = sum(r["h_v2g_incentives"] for r in successful_routes) / len(successful_routes)

            # Calculate energy difference
            energy_diff_pct = ((avg_h_energy - avg_e_energy) / avg_e_energy * 100) if avg_e_energy > 0 else 0
            green_diff = avg_h_green - avg_e_green

            print(f"\n{Fore.CYAN}===== SUMMARY STATISTICS ====={Style.RESET_ALL}")
            print(f"Number of successful routes: {len(successful_routes)}/{len(comparison_results)}")

            print(f"\n{Fore.GREEN}Energy Consumption (kWh):{Style.RESET_ALL}")
            print(f"  Heuristic: {avg_h_energy:.2f}")
            print(f"  Eco-Routing: {avg_e_energy:.2f}")

            if energy_diff_pct < 0:
                print(f"  {Fore.BLUE}Heuristic uses {abs(energy_diff_pct):.2f}% LESS energy than Eco-Routing{Style.RESET_ALL}")
            else:
                print(f"  {Fore.RED}Heuristic uses {energy_diff_pct:.2f}% MORE energy than Eco-Routing{Style.RESET_ALL}")

            print(f"\n{Fore.GREEN}Green Zone Coverage (%):{Style.RESET_ALL}")
            print(f"  Heuristic: {avg_h_green:.2f}%")
            print(f"  Eco-Routing: {avg_e_green:.2f}%")

            if green_diff > 0:
                print(f"  {Fore.BLUE}Heuristic has {green_diff:.2f} percentage points higher green zone coverage{Style.RESET_ALL}")
            else:
                print(f"  {Fore.RED}Heuristic has {abs(green_diff):.2f} percentage points lower green zone coverage{Style.RESET_ALL}")

            print(f"\n{Fore.GREEN}V2G Incentives:{Style.RESET_ALL}")
            print(f"  Average: ${avg_v2g:.2f}")

    return road_network, comparison_results
```

```python
def visualize_paths(graph, paths_dict, figsize=(12, 10)):
    """
    Visualize multiple paths on the same graph with green zones and charging stations.

    Parameters:
        graph (nx.Graph): Road network graph
        paths_dict (dict): Dictionary mapping path names to paths (lists of nodes)
        figsize (tuple): Figure size
    """
    import matplotlib.pyplot as plt
    import networkx as nx
    from matplotlib.lines import Line2D
    import matplotlib.patches as patches
    import numpy as np

    plt.figure(figsize=figsize)

    # Get node positions
    pos = nx.get_node_attributes(graph, 'pos')

    # Define color map for different paths
    colors = ['blue', 'green', 'red', 'magenta', 'cyan', 'yellow', 'black']
    linestyles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1)), (0, (5, 1)), (0, (3, 5, 1, 5))]

    # Draw the network
    nx.draw_networkx_edges(graph, pos, alpha=0.1, width=0.5)

    # Draw green zones if available
    ax = plt.gca()  # Get current axis to add patches
    if 'green_zones' in graph.graph:
        for zone in graph.graph['green_zones']:
            center = zone['center']
            radius = zone['radius']
            circle = plt.Circle(center, radius, color='green', alpha=0.2,
                              fill=True, label='Green Zone')
            ax.add_patch(circle)

    # Draw charging stations
    charging_nodes = [node for node in graph.nodes()
                     if graph.nodes[node].get('charging_station', False)]
    nx.draw_networkx_nodes(graph, pos, nodelist=charging_nodes,
                         node_color='orange', node_shape='s', node_size=100,
                         label='Charging Station')

    # Draw paths
    legend_elements = []
    for i, (path_name, path) in enumerate(paths_dict.items()):
        path_color = colors[i % len(colors)]
        path_style = linestyles[i % len(linestyles)]

        # Create edges from path
        path_edges = [(path[j], path[j+1]) for j in range(len(path)-1)]

        # Draw the path
        nx.draw_networkx_edges(graph, pos, edgelist=path_edges,
                             width=2.5, edge_color=path_color, style=path_style)

        # Add to legend
        legend_elements.append(Line2D([0], [0], color=path_color, lw=2.5,
                                    linestyle=path_style, label=path_name))

    # Draw origin and destination
    if paths_dict and list(paths_dict.values()):
        first_path = list(paths_dict.values())[0]
        if first_path:
            origin = first_path[0]
            destination = first_path[-1]

            nx.draw_networkx_nodes(graph, pos, nodelist=[origin],
                                 node_color='green', node_shape='o', node_size=150,
                                 label='Origin')
            nx.draw_networkx_nodes(graph, pos, nodelist=[destination],
                                 node_color='red', node_shape='o', node_size=150,
                                 label='Destination')

            # Add origin/destination to legend
            legend_elements.append(Line2D([0], [0], marker='o', color='w', markerfacecolor='green',
                                       markersize=10, label='Origin'))
            legend_elements.append(Line2D([0], [0], marker='o', color='w', markerfacecolor='red',
                                       markersize=10, label='Destination'))

    # Add charging stations to legend
    legend_elements.append(Line2D([0], [0], marker='s', color='w', markerfacecolor='orange',
                                markersize=10, label='Charging Station'))

    # Add green zone to legend if applicable
    if 'green_zones' in graph.graph and graph.graph['green_zones']:
        legend_elements.append(patches.Patch(facecolor='green', alpha=0.2,
                                          edgecolor='green', label='Green Zone'))

    # Add legend
    plt.legend(handles=legend_elements, loc='best')

    # Remove axis
    plt.axis('off')

    plt.title('Route Comparison')
    plt.tight_layout()

    return plt.gca()
```

```python
def visualize_route_comparison(graph, origin, destination, vehicle_params=None):
    """
    Visualize and compare routes from different algorithms for a single OD pair.

    Parameters:
        graph (nx.Graph): Road network graph
        origin (int): Source node ID
        destination (int): Target node ID
        vehicle_params (dict, optional): Vehicle parameters

    Returns:
        None (displays plots and statistics)
    """
    import matplotlib.pyplot as plt
    import pandas as pd
    import networkx as nx
    from tabulate import tabulate
    from colorama import Fore, Style

    if vehicle_params is None:
        # Use default vehicle parameters if none provided
        vehicle_params = VEHICLE_PARAMETERS

    print(f"{Fore.CYAN}⚡ Route Comparison: {origin} → {destination} ⚡{Style.RESET_ALL}")

    try:
        # Run all algorithms
        h_path, h_energy, h_metrics = heuristic_routing_algorithm(graph, origin, destination, vehicle_params)
        e_path, e_energy, e_metrics = eco_routing_algorithm(graph, origin, destination, vehicle_params)
        s_path, s_energy, s_metrics = shortest_path_distance(graph, origin, destination, vehicle_params)
        f_path, f_energy, f_metrics = fastest_path_time(graph, origin, destination, vehicle_params)

        # Create paths dictionary for visualization
        paths_dict = {
            'Heuristic': h_path,
            'Eco-Routing': e_path,
            'Shortest Path': s_path,
            'Fastest Path': f_path
        }

        # Visualize the routes
        visualize_paths(graph, paths_dict, figsize=(12, 6))
        plt.show()

        # Create comparison table
        compare_data = {
            'Algorithm': ['Heuristic', 'Eco-Routing', 'Shortest Path', 'Fastest Path'],
            'Energy (kWh)': [h_energy, e_energy, s_energy, f_energy],
            'Time (s)': [h_metrics['time'], e_metrics['time'], s_metrics['time'], f_metrics['time']],
            'Distance (m)': [h_metrics['distance'], e_metrics['distance'], s_metrics['distance'], f_metrics['distance']],
            'Total Cost ($)': [h_metrics['total_cost'], e_metrics['total_cost'], s_metrics['total_cost'], f_metrics['total_cost']]
        }

        compare_df = pd.DataFrame(compare_data)

        # Print route comparison
        print("\nRoute Comparison:")
        print(tabulate(compare_df, headers='keys', tablefmt='pretty', showindex=False))

        # Print heuristic-specific metrics
        print(f"\n{Fore.GREEN}Heuristic Algorithm Specific Metrics:{Style.RESET_ALL}")
        print(f"  V2G Incentives: ${h_metrics['v2g_incentives']:.2f}")
        print(f"  Green Zone Penalties: ${h_metrics['green_zone_penalties']:.2f}")
        print(f"  Charging Stations Visited: {len(h_metrics['charging_stations_visited'])}")

        if len(h_metrics['charging_stations_visited']) > 0:
            print(f"  Charging Stations: {h_metrics['charging_stations_visited']}")

        # Calculate improvement percentages
        h_vs_e_energy = ((h_energy - e_energy) / e_energy * 100) if e_energy > 0 else 0

        print(f"\n{Fore.YELLOW}Energy Consumption Comparison:{Style.RESET_ALL}")
        if h_vs_e_energy < 0:
            print(f"  {Fore.GREEN}Heuristic uses {abs(h_vs_e_energy):.2f}% LESS energy than Eco-Routing{Style.RESET_ALL}")
        else:
            print(f"  {Fore.RED}Heuristic uses {h_vs_e_energy:.2f}% MORE energy than Eco-Routing{Style.RESET_ALL}")

    except Exception as e:
        print(f"{Fore.RED}Error visualizing route comparison: {str(e)}{Style.RESET_ALL}")
```
